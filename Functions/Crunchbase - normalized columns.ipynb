```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 100
```

### Load data and add total investment column

```python
df_combine = pd.read_csv('dfcombine.csv', encoding='latin1')
df_combine['total_investment'] = (df_combine.funded_object_status_acquired + 
                                    df_combine.funded_object_status_closed +
                                    df_combine.funded_object_status_ipo +
                                    df_combine.funded_object_status_operating)
```

### Create lists of column names for startup categories

```python
catlist = list(df_combine.iloc[:,9:51].columns) #list of all startup category columns
#statuslist = list(df_combine.iloc[:,5:9].columns)
#roundlist = list(df_combine.iloc[:,51:60].columns)
print(catlist[:10]) #make sure this starts at advertising and ends with web
```

#### Create normalized columns for categories

```python
normalizednames = []

def newcols(columnlist):
    n = 0
    for i in columnlist:
        normalizednames.append(columnlist[n][16:] + '_normalized')
        #if you want to recreate this for statuslist must change the above range from 16: to 21:
        #if you want to recreate this for roundlist must change the above range from 16: to 19:
        n += 1
    normalizednames
    
newcols(catlist) #must change to statuslist or roundlist if updating

normalizednames
```

```python
def normalizedstatus(column):
    n=0
    for i in column:
        df_combine[str(normalizednames[n])] = (df_combine[i]/df_combine.total_investment)*1.0
        n += 1   
        

normalizedstatus(catlist) #update to statuslist or roundlist if changing
```

### Create normalized columns for degrees and investments

```python
df_combine['degree_normalized'] = ((df_combine.degree-df_combine.degree.mean())/df_combine.degree.std())*1.0
df_combine['investments_normalized'] = ((df_combine.total_investment-df_combine.total_investment.mean())/df_combine.total_investment.std())*1.0
```

```python
print 'investments normalized max: %f' %df_combine.investments_normalized.max()
print 'investments normalized min: %f' %df_combine.investments_normalized.min()
print 'degrees normalized max: %f' %df_combine.degree_normalized.max()
print 'degrees normalized min: %f' %df_combine.degree_normalized.min()
```
